```python
#y = w1x1+w2x2
x1 = 8
x2 = 6
#w1 = 3
#w2 = 2
#ya = (w1*x1)+(w2*x2)
ya = 36
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 0.002
w1 = np.random.randint(1,10)
w2 = np.random.randint(1,10)
w1_list = []
w2_list = []
error_list = []
ypred_list = []
for i in range(20):
    w1_list.append(w1)
    w2_list.append(w2)
    #phase 1 : random weight selection and feed forward
    ypred = np.round((w1*x1)+(w2*x2),decimals = 2)
    ypred_list.append(ypred)
    #phase 2 : error calculation
    error = (ya-ypred)**2
    error_list.append(error)
    #phase 3 : backward propagation
    grad1 = 2*(ya-ypred)*(-x1)
    grad2 = 2*(ya-ypred)*(-x2)
    #phase 4 : weight update
    w1 = w1 - (alpha*grad1)
    w2 = w2 - (alpha*grad2)
print(f'The predicted value is {ypred} while the actual value is {ya} and the error is {error}')
```

```python
plt.plot(error_list)
plt.title('Error Graph')
plt.xticks(range(1,20))
plt.xlabel('Epochs')
plt.ylabel('Error value')
plt.show()
```

```python
plt.plot(ypred_list)
plt.title('Predicting Output')
plt.xticks(range(1,20))
plt.xlabel('Epochs')
plt.ylabel('Output value')
plt.show()
```
