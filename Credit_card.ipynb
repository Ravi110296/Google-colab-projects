```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
```

```python
Normalized_Amount_column = ss.fit_transform(data['Amount'].values.reshape(-1,1))
Normalized_Amount_column.shape
```

```python
x = data.drop('Class',axis = 1)
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.3, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
```

```python
dt.fit(xtrain,ytrain)
```

```python
rf.fit(xtrain,ytrain)
```

```python
ypred_dt = dt.predict(xtest)
ypred_rf = rf.predict(xtest)
ypred_dt, ypred_rf
ypred = pd.DataFrame({'dt':ypred_dt, 'rf':ypred_rf})
sns.countplot(ypred)
plt.show()
```

```python
print(rf.score(xtest,ytest))
print(dt.score(xtest,ytest))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred_dt,labels=ytest.value_counts().index))
```

```python
print(classification_report(ytest,ypred_rf,labels=ytest.value_counts().index))
```

```python
# as per the performance matrix Decision tree was able to detect 82 percent of the 16 frauds, while the random forest was able to predict 93 percent times
# We have taken precision here as False Positive ( it was a fraud but model considered it as not fraud ) is more crucial for this problem
# Hence Random Forest better suits this problem
```
