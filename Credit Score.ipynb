```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/credit_score.csv')
```

```python
data.head()
```

```python
data.drop(['ID','Customer_ID','Name','SSN', 'Type_of_Loan','Credit_History_Age'], axis = 1,inplace = True)
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
for i in data.columns:
  if data[i].isnull().sum() !=0:
    print(data[i].value_counts())
```

```python
data["Age"]=data["Age"].str.replace("_","")
data["Age"]=data["Age"].astype (int)
data["Occupation"]=data["Occupation"].replace("______", np.nan)
data["Annual_Income"]=data["Annual_Income"].str.replace("_","")
data["Annual_Income"]=data["Annual_Income"].astype(float)
data["Num_of_Loan"]=data["Num_of_Loan"].str.replace("_","")
data["Num_of_Loan"]=data["Num_of_Loan"].astype(int)
data["Num_of_Delayed_Payment"]=data["Num_of_Delayed_Payment"].str.replace("_","")
data["Num_of_Delayed_Payment"]=data["Num_of_Delayed_Payment"].astype(float)
data["Credit Score"]=data["Credit_Score"].replace(["Poor", "Standard", "Good"],[0,1,2])
data["Monthly_Balance"]=data["Monthly_Balance"].str.replace("_","")
data["Monthly_Balance"]=data["Monthly_Balance"].astype(float)
data["Payment_Behaviour"]=data["Payment_Behaviour"].replace("! @9#%8", np.nan)
data["Amount_invested_monthly"]=data["Amount_invested_monthly"].str.replace("_","")
data["Amount_invested_monthly"]=data["Amount_invested_monthly"].astype (float)
data["Payment_of_Min_Amount"]=data["Payment_of_Min_Amount"].replace("NM", "No")
data["Payment_of_Min_Amount"]=data["Payment_of_Min_Amount"].replace(["Yes","No"],[1,0])
data["Outstanding_Debt"]= data["Outstanding_Debt"].str.replace("_","")
data["Outstanding_Debt"]=data["Outstanding_Debt"].astype(float)
data["Credit_Mix"]=data["Credit_Mix"].replace("_", np.nan)
data["Credit_Mix"]=data["Credit_Mix"].replace(["Standard", "Good", "Bad"],[1,2,0])
data["Changed_Credit_Limit"]= data["Changed_Credit_Limit"].replace("_", np.nan)
data["Changed_Credit_Limit"]=data["Changed_Credit_Limit"].astype(float)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in data.columns:
  if data[i].dtypes=='object':
    data[i] = le.fit_transform(data[i])
```

```python
data.info()
```

```python
data.drop('Credit Score',axis = 1, inplace = True)
```

```python
x = data.drop('Credit_Score',axis = 1)
y = data['Credit_Score']
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
xtrain = ss.fit_transform(xtrain)
xtest = ss.transform(xtest)
```

```python
xtrain.shape,ytrain.shape
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
VIF = pd.DataFrame()
VIF['features'] = x.columns
VIF['VIF'] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
VIF
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=200)
```

```python
rf.fit(xtrain,ytrain)
```

```python
ypreed_rf = rf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ypreed_rf,ytest)
```

```python
param_grid = {
    'n_estimators':[100,200],
    'criterion':['gini','entropy'],
    'max_depth':[None,5,10],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = RandomForestClassifier()
```

```python
grid = GridSearchCV(model,param_grid,cv=2)
```

```python
grid.fit(xtrain,ytrain)
```

```python
best_params = grid.best_params_
best = grid.best_estimator_
best_params,best
```

```python
ypred_hpt = best.predict(xtest)
```

```python
accuracy_score(ypred_hpt,ytest)
```
