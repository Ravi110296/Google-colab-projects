```python
import transformers
```

```python
classifier = transformers.pipeline('sentiment-analysis',model = 'nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
classifier('Abrazar la cara es una de las mejores, si no la mejor, biblioteca.')
```

```python
classifier.tokenizer.get_vocab()
```

```python
classifier('hugging face is one of the best if not the best library')
```

```python
classifier('ハグフェイスは最高ではないにしても最高のライブラリの1つです')
```

```python
results = classifier(['pizza combined with cola will be amazing','ベジタリアンが日本で生きていくのは難しい'])
for result in results:
  print(result)
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification, pipeline
```

```python
model_name = 'nlptown/bert-base-multilingual-uncased-sentiment'
token = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForSequenceClassification.from_pretrained(model_name)
classif = pipeline('sentiment-analysis',model = model, tokenizer = token)
```

```python
classif(['pizza combined with cola will be terrific','日本はシーフードで有名です'])
```
