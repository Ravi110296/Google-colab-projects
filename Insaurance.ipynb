```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
insaurance['sex'] = le.fit_transform(insaurance['sex'])
insaurance.smoker = le.fit_transform(insaurance.smoker)
insaurance.region = le.fit_transform(insaurance.region)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = insaurance.drop(['age','charges', 'bmi', 'children','region','sex','smoker'], axis = 1)
y = insaurance['charges']
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
xtrain = sc.fit_transform(xtrain)
xtest = sc.fit_transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
ypred = lr.predict(xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(ytest,ypred)
r2
```
