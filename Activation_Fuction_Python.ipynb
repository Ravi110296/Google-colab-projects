```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##defining the functions used in Neural network
def sigmoid(x):
  sig = 1/(1+np.exp(-x))
  return sig

def tanh(x):
  tanh = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
  return tanh

def relu(x):
  relu = []
  for i in x:
    relu.append(max(0,i))
  return relu
```

```python
l = np.arange(-10,11)
```

```python
plt.plot(l,sigmoid(l))
plt.vlines(x = 0,ymin = 0,ymax = 1,color = 'red')
plt.hlines(y = 0,xmin = -10,xmax = 10,color = 'red')
plt.title('Sigmoid(x)')
plt.show()
```

```python
plt.plot(l,tanh(l))
plt.hlines(y = 0, xmin = -10, xmax = 10, color = 'red')
plt.vlines(x = 0, ymin = -1, ymax = 1,color = 'red')
plt.title('tanh(x)')
plt.show()
```

```python
plt.plot(l,relu(l))
plt.hlines(y = 0, xmin = -10, xmax = 10, color = 'red')
plt.vlines(x = 0, ymin = 0, ymax = 10, color = 'red')
plt.title('Relu(x)')
plt.show()
```

```python
## actual Data
x1 = 3
x2 = 2
yact = 17
#equation = 3x1+4x2
```

```python
lr = 0.01
w1 = np.random.randint(1,10)
w2 = np.random.randint(1,10)
err = []
pred = []
for i in range(0,20):
  ypred = w1*x1+w2*x2
  pred.append(ypred)
  error = (yact-ypred)**2
  err.append(error)
  grad1 = 2*(yact-ypred)*-x1
  grad2 = 2*(yact-ypred)*-x2
  w1 = w1-lr*grad1
  w2 = w2-lr*grad2
```

```python
print(ypred,error)
```

```python
epochs = np.arange(1,21)
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.plot(epochs,pred)
plt.title('Ypred vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Ypred')
plt.subplot(1,2,2)
plt.plot(epochs,err)
plt.title('Error vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.show()
```
