```python
ya = np.array([0,0,1,0,0])
```

```python
ypred = np.array([0.09,0.09,0.64,0.1,0.08])
ypred
```

```python
mse = ((ya-ypred)**2)/10
mse
```

```python
plt.plot(range(0,5),mse)
```

```python
w_log = ya*ypred+(1-ya)*(1-ypred)
w_log
```

```python
plt.plot(range(0,5),ya*ypred+(1-ya)*(1-ypred))
```
